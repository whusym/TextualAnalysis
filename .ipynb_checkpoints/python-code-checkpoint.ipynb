```python
d = hobbes.word_counts
```

```python
#Adding more stopwords. I need an aggressive way of stopwords.
# nltk.download('stopwords')  Not necessary if you have done it once
from nltk.corpus import stopwords
stop_words = list(set(stopwords.words('english')))
with open('stopwords', 'r') as myfile:
    sw = [i.strip().split(' ') for i in myfile]
sw1 = [val.lower() for sublist in sw for val in sublist]
stop_words.extend(sw1)
stop_words = set(stop_words)
# print (len(stop_words))
```

```python


d_sorted = dict(sorted(d.items(), key=lambda k_v: k_v[1], reverse=True))

# print (d_sorted)
d_sorted_copy = dict(d_sorted)
for key, value in d_sorted_copy.items():
    if key in stop_words:
        del d_sorted[key]
    elif len(str(key)) == 1:
        del d_sorted[key]
```

```python
# print (d_sorted)
d_sorted = sorted(d_sorted.items(), key=lambda k_v: k_v[1], reverse=True)
```

```python
# print (dict(d_sorted))
d_sorted_plot = dict(d_sorted[0:60])
```

```python

plt.bar(range(len(d_sorted_plot)), sorted(d_sorted_plot.values(), reverse=True), align='center')
plt.xticks(range(len(d_sorted_plot)), d_sorted_plot.keys(), fontsize = 28, rotation=85)
# plt.figure(num=None, figsize=(20,12), dpi=80, facecolor='w', edgecolor='k')
plt.rcParams["figure.figsize"] = (32,24)
plt.show()
```
